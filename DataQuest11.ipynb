### Importing Libraries

```python
import pandas as pd
import datetime
```

### Reading Data

```python
df2020 = pd.read_excel('data_2020.xlsx')
df2021 = pd.read_excel('data_2021.xlsx')
df2022 = pd.read_excel('data_2022.xlsx',skiprows = 1)
```

### PreProcessing

```python
df2021 = df2021.drop(0).reset_index(drop=True)
```

```python
df2020.columns = df2021.columns
df2022.columns = df2021.columns
```

```python
def changeTypeDate(value):
    form = "%d/%m/%Y"
    try:
        return pd.to_datetime(value,errors = 'coerce',format =form)
    except:
        pd.NaN

df2022['DATE'] = df2022['DATE'].apply(changeTypeDate)
df2020['DATE'] = df2020['DATE'].apply(changeTypeDate)
```

```python
l_date= ""
format = "%d/%m/%Y"
df2020['DATE'] = df2020['DATE'].fillna(0)
for index in range(0,len(df2020)):
    if df2020['DATE'][index]==0:
        df2020['DATE'][index] = datetime.datetime.strptime(l_date, format)
    else:
        l_date = df2020['DATE'][index].strftime(format)
```

```python
last_date = '01/01/2022'
format = "%d/%m/%Y"
df2022['DATE'] = df2022['DATE'].fillna(0)
for index in range(0,len(df2022)):
    if df2022['DATE'][index] == 0:
        df2022['DATE'][index] = datetime.datetime.strptime(last_date, format)
    else:
        last_date = df2022['DATE'][index].strftime(format)
```

### PreProcessing 2021

```python
temperature = df2021["Temperature"].tolist()

cleaned_temperature_list = []
for temp in temperature:
    if isinstance(temp, str):
        cleaned_temperature_list.append(temp.replace('\xa0°F', ''))
    elif isinstance(temp, float):
        cleaned_temperature_list.append(str(temp))

        

df2021["Temperature"]=cleaned_temperature_list
```

```python
windspeed = df2021["Wind Speed"].tolist()
windsspeed_list = [str(speed).replace('mph', '') for speed in windspeed]
df2021["Wind Speed"] = windsspeed_list
```

```python
Wind_Gust = df2021["Wind Gust"].tolist()
Wind_Gust_list= [str(s).replace('mph', '') for s in Wind_Gust]
df2021["Wind Gust"] = Wind_Gust_list
```

```python
Pressure = df2021["Pressure"].tolist()
Pressure_list= [str(s).replace('mph', '') for s in Wind_Gust]
df2021["Pressure"] = Pressure_list
```

```python
Precip = df2021["Precip."].tolist()
Precip_list = [str(s).replace('in', '') for s in Precip]
df2021["Precip."] = Precip_list
```

```python
weather_mapping = {
    'Fog': 1,
    'Smoke': 2,
}

condition = df2021["Condition"].tolist()
condition_list = [weather_mapping.get(c, 0) for c in condition]
df2021["Condition"] = condition_list
```

```python
def changeNumeric(value):
    return([int(value[0]),int(value[0]),int(value[1]),int(value[2]),int(value[3]),int(value[4]),int(value[5]),int(value[6])])
    
cols_change = ['Temperature', 'Dew Point', 'Humidity', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']
df2020[cols_change] = df2020[cols_change].apply(changeNumeric)
df2021[cols_change] = df2021[cols_change].apply(changeNumeric)
df2022[cols_change] = df2022[cols_change].apply(changeNumeric)
```

```python
columns_to_convert = ['Temperature', 'Dew Point', 'Humidity', 'Wind', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']
df2020[columns_to_convert] = df2020[columns_to_convert].apply(pd.to_numeric, errors='coerce')
columns_to_convert_to = ['Temperature', 'Dew Point', 'Humidity', 'Wind', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']
df2021[columns_to_convert] = df2021[columns_to_convert_to].apply(pd.to_numeric, errors='coerce')
columns_to_convert_too = ['Temperature', 'Dew Point', 'Humidity', 'Wind', 'Wind Speed', 'Wind Gust', 'Pressure', 'Precip.']
df2022[columns_to_convert] = df2022[columns_to_convert_too].apply(pd.to_numeric, errors='coerce')
```
