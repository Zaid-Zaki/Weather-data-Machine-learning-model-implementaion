```python
import pandas as pd

# Read data from Excel files
dataframe_2022 = pd.read_excel("data_2022.xlsx", skiprows=1)
dataframe_2021 = pd.read_excel("data_2021.xlsx")
dataframe_2020 = pd.read_excel("data_2020.xlsx")

# Remove null values from each DataFrame
dataframe_2022.dropna(inplace=True)
dataframe_2021.dropna(inplace=True)
dataframe_2020.dropna(inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data loaded into a DataFrame called dataframe_2022
# If not, load your data using pd.read_excel() or any other appropriate method
# dataframe_2022 = pd.read_excel("data_2022.xlsx")

# Convert 'Time' column to string representation
dataframe_2022['Time'] = dataframe_2022['Time'].astype(str)

# Plot Temperature over Time
plt.figure(figsize=(10, 6))
plt.plot(dataframe_2022['Time'], dataframe_2022['Temperature'], color='blue')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data loaded into a DataFrame called dataframe_2022
# If not, load your data using pd.read_excel() or any other appropriate method
# dataframe_2022 = pd.read_excel("data_2022.xlsx")

# Convert 'Time' column to string representation
dataframe_2022['Time'] = dataframe_2022['Time'].astype(str)

# Plot Wind Speed over Time
plt.figure(figsize=(10, 6))
plt.plot(dataframe_2022['Time'], dataframe_2022['Wind Speed'], color='blue')
plt.xlabel('Time')
plt.ylabel('Wind Speed (mph)')
plt.title('Wind Speed Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Humidity over Time
plt.figure(figsize=(10, 6))
plt.plot(dataframe_2022['Time'], dataframe_2022['Humidity'], color='green')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')
plt.title('Humidity Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Pressure over Time
plt.figure(figsize=(10, 6))
plt.plot(dataframe_2022['Time'], dataframe_2022['Pressure'], color='red')
plt.xlabel('Time')
plt.ylabel('Pressure (in)')
plt.title('Pressure Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Precipitation over Time
plt.figure(figsize=(10, 6))
plt.plot(dataframe_2022['Time'], dataframe_2022['Precip.'], color='purple')
plt.xlabel('Time')
plt.ylabel('Precipitation (in)')
plt.title('Precipitation Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
